```python
import pandas as pd
import numpy as np
import prepare as p

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

from env import get_connection
```

# Take the work we did in the lessons further:

# 1. What other types of models (i.e. different classifcation algorithms) could you use?

```python
url = get_connection('spam_db')

query = 'SELECT * FROM spam'

df = pd.read_sql(query, url, index_col='id')

df.head()
```

```python


df['clean_text'] = df.text.apply(p.advanced_clean, extra_words=['us'])
```

```python
df.head()
```

```python
df['lemmas'] = df.text.apply(p.advanced_clean, extra_words=['us'], l = True)
```

```python
df.head()
```

```python
words = pd.Series(df.lemmas.str.split())
words.value_counts().sum()
```

```python
tf = (pd.DataFrame({'raw_count': words.value_counts()})
 .assign(frequency=lambda df: df.raw_count / df.raw_count.sum())
 .assign(augmented_frequency=lambda df: df.frequency / df.frequency.max()))
```

```python
tf.head()
```

```python
tf.frequency.max()
```

```python
X = df.lemmas
y = df.label

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7,
                                                    random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.value_counts()
```

```python
 y_train.shape[0] * 100
```

# Baseline

```python
baseline_acc = y_train.value_counts().max() / y_train.shape[0] * 100

print(f'Baseline accuracy: {round(baseline_acc, 2)}.')
```

```python
tfidf = TfidfVectorizer()

X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

X_train_tfidf[:10]
```

```python
pd.DataFrame(X_train_tfidf.todense(), columns=tfidf.get_feature_names_out())
```

# Logistic Regression

```python
lm = LogisticRegression(class_weight = 'balanced')

lm.fit(X_train_tfidf, y_train)
```

```python
y_train_res = pd.DataFrame({'actual': y_train,
                            'preds': lm.predict(X_train_tfidf)})

y_train_res.head()
```

```python
print(classification_report(y_train_res.actual, y_train_res.preds))
```

```python
y_test_res = pd.DataFrame({'actual': y_test,
                           'preds': lm.predict(X_test_tfidf)})

y_test_res.head()
```

```python
print(classification_report(y_test_res.actual, y_test_res.preds))
```

# Random Forest

```python
rf = RandomForestClassifier(class_weight = 'balanced')
rf.fit(X_train_tfidf, y_train)
```

```python
y_train_res = pd.DataFrame({'actual': y_train,
                            'preds': rf.predict(X_train_tfidf)})

y_train_res.head()
```

```python
print(classification_report(y_train_res.actual, y_train_res.preds))
```

```python
y_test_res = pd.DataFrame({'actual': y_test,
                           'preds': rf.predict(X_test_tfidf)})

y_test_res.head()
```

```python
print(classification_report(y_test_res.actual, y_test_res.preds))
```

# Decision Tree

```python
dt = DecisionTreeClassifier(class_weight = 'balanced')
    
dt.fit(X_train_tfidf, y_train)
```
